```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
dataset_path = '/content/Housing.csv'
df = pd.read_csv(dataset_path)
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns.to_list()
print(categorical_cols)
```

```python
ordinal_encoder = OrdinalEncoder()
encoded_catgorical_cols = ordinal_encoder.fit_transform(df[categorical_cols])
encoded_categorical_df = pd.DataFrame(encoded_catgorical_cols, columns=categorical_cols)
numerical_df = df.drop(categorical_cols, axis=1)
encoded_df = pd.concat([encoded_categorical_df, numerical_df], axis=1)
```

```python
normalizer = StandardScaler()
dataset_arr = normalizer.fit_transform(encoded_df)
```

```python
X, y = dataset_arr[:, 1:], dataset_arr[:, 0]
```

```python
test_size = 0.3
random_state = 1
is_shuffle = True
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state, shuffle=is_shuffle)
```

```python
regressor = DecisionTreeRegressor(random_state=random_state)
regressor.fit(X_train, y_train)
```

```python
regressor = AdaBoostRegressor(random_state=random_state)
regressor.fit(X_train, y_train)
```

```python
regressor = GradientBoostingRegressor(random_state=random_state)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
```
